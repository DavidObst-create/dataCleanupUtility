```python
def multithread_process(df, worker_task, num_workers=None, num_chunks=200): 
    if num_workers=None:
        try:
            num_workers=cpu_count()
        except Exception as e:
            print("Unable to detect CPU. Setting workers to 1")
            num_workers=1
    
    chunks = np.array_split(df, num_chunks)
    with Pool(num_workers) as p: 
    results = list(
        tqdm(
            p.imap(worker_task, chunks),
            total=num_chunks,
            desc="Filtering rows"
        )
    )
    return pd.concat(results, ignore_index=True)
    
```
